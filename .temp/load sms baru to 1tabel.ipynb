```python
import pandas as pd
import re
```

```python
import os

path = r'SMS' #direktori letak files

files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if '.csv' in file:
            files.append(os.path.join(r, file))

# for f in files:
#     print(f)
print(len(files))
sms = list()
for al in files:
#     print(al)
    foo = pd.read_csv(al)
    msg = foo.Received.tolist()[1:]
    for m in msg:
        m = str(m).rstrip()
        m = m.replace('\r',' ').replace('\n',' ')
        m = re.sub(' +', ' ', m.lstrip().rstrip())
        sms.append(m)
sms = list(set(sms))
print('jumlah sms',len(sms))
dict_ = {
    'sms':sms
}
df = pd.DataFrame.from_dict(dict_)
df
```

```python
df.to_excel("sms baru/new_sms.xlsx")
```
