```python
import pandas as pd
```

```python
data = pd.read_excel("hasil/fitur (version 2).xlsx")
fitur = data['fitur'].tolist()
kode = data['Unnamed: 2'].tolist()
data.head()
```

```python
import joblib
hapus_ = list()
ganti_ = dict()
for i, j in zip(fitur, kode):
#     print(i,j, type(j))
    
    if type(j) is not float:
        i = str(i)
        j = str(j)
        if j == "1":
            hapus_.append(i)
        else:
            ganti_.update({i:j})
#             print(i,j)
dict_={
    "hapus":hapus_,
    "ganti":ganti_
}

dict_['hapus']
```
