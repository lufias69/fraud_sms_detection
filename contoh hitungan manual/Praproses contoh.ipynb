```python
data = pd.read_excel("contoh_perhitungan_new.xlsx", sheet_name = 'Sheet1')
sms = data.sms.tolist()
label = data.label.tolist()
data
```

# Praproses

```python
import re
casfolding_list = list()
token_list = list()
normal_list = list()
stemming_list = list()
stopwors_list = list()
data_bersih = list()

ngram_list = list()

for ix, i in enumerate(sms):
    i = i.replace("..", ".")
    i = i.lower()
    casfolding_list.append(i)
    
    #tokenisasi
    i = tokenize.tokenisasi(i)
    token_list.append(i)
    i = " ".join(i)
    i = re.sub(' +', ' ',i)
#     i = pps.preprocessing(i)
#     print(i)
    
    #normalisasi
    i = i.replace(' dg ', ' dengan ').replace(' dpt ', ' dapat ').replace(' sama2 ', '').replace(' yg ', ' yang ').replace(' tp ', ' tapi ').replace(' rb', ' ribu ')
    i = i.replace(" temen2 ", " teman ").replace("temen2 ", " teman ").replace(" ku ", "").replace(" hmm", "").replace(" sih ", "").replace("2 5 ", "").replace('001', "").replace('0100', "").replace('081281108799', "").replace('10', "").replace('100', "").replace('1572', "").replace('2011', "").replace('2015', "").replace('35', "").replace('503', "").replace('550', "").replace('600', "").replace('700', "").replace('905', "").replace('9997175', "").replace('30', "").replace('400', "").replace('4g', "").replace('91', "")
    i = i.replace(' kt ', ' kita ').replace(' bpk ', ' bapak ').replace(' utk ', ' untuk ').replace(' dri ', ' dari ').replace('!', ' ')
    i = re.sub(' +', ' ',i)
#     i = re.sub('.+',// '.',i)
#     i = re.sub('.,', ',',i)
    normal_list.append(i)
    
    #Stemming
    i = stm.stemmer_kata(i)
    stemming_list.append(i)
#     print(i)
    #gram
#     i_ng = " ".join(tokenize.tokenisasi(i))
    i = re.sub(' +', ' ',i).strip()
    i_ng = ng.ngramku(i, delimiter=delimiter, n=4)['string']
    ngram_list.append(i_ng)
    
    #stopwords
    i = stm.stop_word(pps.preprocessing(i))
    stopwors_list.append(i)
    data_bersih.append(i)
    
dictt = {
    "data_asli":sms,
    'casfolding_list':casfolding_list,
    'token_list':token_list,
    'normal_list':normal_list,
    'stemming_list':stemming_list,
    "ngram_list":ngram_list,
    'stopwors_list':stopwors_list,
}
dictt.update({"sms_bersih":data_bersih})
x = pd.DataFrame.from_dict(dictt)
# x.to_excel("contoh_data_sms_bersih_all2.xlsx")
x
```
