```python
from modulku import praproses as pps, StemNstopW as stm
from joblib import load
from replace import normalisasi as nrm
import pandas as pd
```

```python
#load Model yang sudah dibuat
tfidf = load('model/tfidf.w')
mnb = load('model/mnb.m')

#praproses
def praproses(sms):
    sms = pps.preprocessing(sms)
    sms = nrm.normalisasi(sms)
    sms = stm.stemmer_kata(sms)
    # sms = stm.stop_word(sms)
    return sms
```

```python
sms = "Isi saldo TCASHmu  melalui jaringan ATM Bersama & ATM BCA! Lakukan transfer antar bank, Pilih bank TCASH (Kode bank 911). Gunakan No HP sbg no rekening tujuan."
```

```python
data = pd.read_excel("../dataset_sms_spam _v1.xlsx")
sms_list = data.Teks.tolist()
data = pd.read_excel("sms_clean_norm_+_new_data.xlsx")
label_asli = data.label.tolist()


data.head()
```

```python
#prediksi
prediksi = list()
for ix, sms in enumerate(sms_list):
    sms = praproses(sms)
    w = tfidf.transform([sms])
    hasil = mnb.predict(w)[0]
    prediksi.append(hasil)
    print("<<",hasil,">>")
    print(sms_list[ix])
    print("="*100)
```

```python
dict_ = {
    "label_asli":label_asli,
    "prediksi":prediksi,
    'sms':sms_list,
    
}
df = pd.DataFrame.from_dict(dict_)

df
```

```python
df.to_excel("hasil prediksi/hasil_prediksi_datalatih.xlsx")
```

```python
benar = 0
salah = 0
ix = 0
for a, b, c in zip(df.label_asli, df.prediksi, df.sms):
    if a== b:
        benar+=1
    else:
        salah +=1
        print(ix, "|", a, "|", b)
        print("                 ", c)
        print("="*100)
    ix+=1
print(benar)
print(salah)
```
